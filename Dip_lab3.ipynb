```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
#task1 Gamma correction
img=cv2.imread('1.jpg')

print('Input')
cv2_imshow(img)

x=5

# s=c*(r**y)
outImg = np.array(255*(img / 255) **x, dtype = 'uint8')
print('Output:')
cv2_imshow(outImg)
```

```python
#task2 min-max contrast stretching

img=cv2.imread('2.jpg')

print('Input:')
cv2_imshow(img)

h,w,ch=img.shape
#print(img.shape)
mx=-1
mn=300
for x in range(h):
  for y in range(w):
    for z in range(ch):
        if mx< img[x][y][z]:
          mx=img[x][y][z]
        if mn> img[x][y][z]:
          mn=img[x][y][z]
        
for x in range(h):
  for y in range(w):
    for z in range(ch):
      img[x][y][z]=255*(img[x][y][z]-mn)/(mx-mn)
print("Output:")
cv2_imshow(img)
```

```python
#task 3 Contrast Stretching+Threshholding
img=cv2.imread('3.jpg')
print('Input:')
cv2_imshow(img)

h,w,ch=img.shape
#print(img.shape)
mx=-1
mn=300
for x in range(h):
  for y in range(w):
    for z in range(ch):
        if mx< img[x][y][z]:
          mx=img[x][y][z]
        if mn> img[x][y][z]:
          mn=img[x][y][z]
        
for x in range(h):
  for y in range(w):
    for z in range(ch):
      img[x][y][z]=255*(img[x][y][z]-mn)/(mx-mn)

print('Output Image1 by Contrast Stretching')
cv2_imshow(img)

gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
h,w=gray.shape
thres1=40
thres2=128
thres3=190
for x in range(h):
  for y in range(w):
      if thres3< gray[x][y]:
    #  if thres2<= gray[x][y]:
          gray[x][y]=255
      elif thres1> gray[x][y] :
      #else:
          gray[x][y]=0

print('Output Image2 by Threshholding')
cv2_imshow(gray)
```

```python
#4th task
#histogram Equilization
img=cv2.imread('4.jpg')

print('Input:')
cv2_imshow(img)

src = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
dst = cv2.equalizeHist(src)
#cv2_imshow(src)
print("Output by Histogram Equalization:")
cv2_imshow(dst)
```
