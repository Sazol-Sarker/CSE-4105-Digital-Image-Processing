```python
import numpy as np
import cv2 
import math
import matplotlib.pyplot as plt
from matplotlib import style
from google.colab.patches import cv2_imshow
```

```python
#task1--erosion
img1=cv2.imread('/content/ball.PNG',0)

print('Input')
cv2_imshow(img1)
print(img1.shape)
h,w=img1.shape
print(img1.shape)
nh=h+2
nw=w+2
print(nh,nw)

mask=[[1,1,1],
       [1,1,1],
       [1,1,1]]

padded_img=a = np.zeros([nh, nw], dtype = int)

for x in range(1,h,1):
  for y in range(1,w,1):
     padded_img[x][y]=img1[x][y]

out_img=a = np.zeros([h+2,w+2], dtype = int)

#print("Padded image ",padded_img)

for i in range(1,h-1,1):
  for j in range(1,w-1,1):
    cnt=0
    for p in range(i-1,i+2):
      for q in range(j-1,j+2):    
         if p==i and q==j:
             cnt+=0
         elif  padded_img[p][q]>0:
               cnt+=1
         
    if cnt==8:
         out_img[i][j]=255
    else: 
         out_img[i][j]=0
   
print("Erosion--Applying 1st Mask:")
cv2_imshow(out_img)

out_img1=a = np.zeros([h+2,w+2], dtype = int)

for i in range(1,h-1,1):
  for j in range(1,w-1,1):
    cnt=0
    for p in range(i-1,i+2):
      for q in range(j-1,j+2):    
         if p==i and q==j:
             cnt+=0
         elif  out_img[p][q]>0:
               cnt+=1
    if cnt==8:
         out_img1[i][j]=255
    else: 
         out_img1[i][j]=0
   
print("Erosion--Applying 2nd Mask:")
cv2_imshow(out_img1)

```

```python
import cv2
import numpy as np

# Reading the input image
img = cv2.imread('/content/ball.PNG', 0)

kernel = np.ones((5, 5), np.uint8)

img_erosion = cv2.erode(img, kernel, iterations=2)
img_dilation = cv2.dilate(img, kernel, iterations=2)

cv2_imshow(img_erosion)
cv2_imshow(img_dilation)
```
